```python
%matplotlib inline
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import math
from scipy.optimize import curve_fit
```

```python
# Importing Data

url = "https://raw.githubusercontent.com/krittikaiitb/KSP2022-selection/master/CepheidData.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
# Cleaning Data

df_cleaned = df
df_cleaned = df_cleaned.drop(df_cleaned[(df_cleaned.V == -99.990)].index)
df_cleaned = df_cleaned.drop(df_cleaned[(df_cleaned.I == -99.990)].index)

df_cleaned.reset_index(inplace=True, drop=True)

df_cleaned.head(10)
```

```python
# Coordinate conversion on test data (first row)
# Results verified from https://www.swift.psu.edu/secure/toop/convert.htm
test = df_cleaned.iloc[0]

# Converting Declination in DMS to Declination in Decimal Degrees (DD)

decl = test.Decl
decl_d, decl_m, decl_s = re.split('[::]', decl)

def dms2deg(d, m, s):
    if(float(d) < 0):
        return (float(d) - float(m)/60 - float(s)/3600)
    else:
        return (float(d) + float(m)/60 + float(s)/3600)

print("Original Declination (in DMS): " + decl)
print("Converted Declination (in DD): " + str(dms2deg(decl_d, decl_m, decl_s)))


# Converting Right Ascension in HMS to Right Ascension in Decimal Degrees (DD)

ra = test.RA
ra_h, ra_m, ra_s = re.split('[::]', ra)

def hms2deg(h, m, s):
    return (15*float(h) + 15*float(m)/60+ 15*float(s)/3600)

print("Original Right Ascension (in HMS): " + ra)
print("Converted Right Ascension (in DD): " + str(hms2deg(ra_h, ra_m, ra_s)))
```

```python
# Coordinate conversion on the whole dataframe

# Converting RA in HMS to RA in DD  
ra = list(df_cleaned["RA"])
ra_dd = [] 

for i in ra:
    ra_h, ra_m, ra_s = re.split('[::]', i)
    ra_dd.append(hms2deg(ra_h, ra_m, ra_s))

df_cleaned["RA_converted(DD)"] = ra_dd

# Converting Decl in DMS to Decl in DD
decl = list(df_cleaned["Decl"])
decl_dd = []

for i in decl:
    decl_d, decl_m, decl_s = re.split('[::]', i)
    decl_dd.append(dms2deg(decl_d, decl_m, decl_s))
    
df_cleaned["Decl_converted(DD)"] = decl_dd
df_cleaned.tail(10)
```

```python
# Plotting RA DEC on Mollweide projection

ra_dd = list(df_cleaned["RA_converted(DD)"]) 
ra_radians = []

for index in ra_dd:
    if (index > 180):
        temp = index - 360
        ra_radians.append(math.radians(temp))
    elif (index < 180):
        ra_radians.append(math.radians(index))

decl_dd = list(df_cleaned["Decl_converted(DD)"]) 
decl_radians = []

for index in decl_dd:
    temp = math.radians(index)
    decl_radians.append(temp)

fig_RAvsDEC = plt.figure(figsize=(10,5))
ax = fig_RAvsDEC.add_subplot(111, projection="mollweide")
ax.scatter(ra_radians, decl_radians, s = 0.1)
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
plt.title("MOLLEWEIDE Projection")
ax.grid(True)
```

```python
# Plotting I and V band magnitudes wrt log (Period)

I = list(df_cleaned["I"])
V = list(df_cleaned["V"])

period = list(df_cleaned["P_1"])
log_period = []

for index in period:
    log_period.append(math.log(index))

fig_BandsvslogP_1, axs = plt.subplots(1, 2, sharey=True)

axs[0].scatter(log_period, I, s=0.1)
axs[0].grid(True)
axs[0].set(xlabel='log(Period)', ylabel='I', title='I vs log(Period)')
axs[0].spines['left'].set_position('zero')
axs[0].spines['right'].set_color('none')
axs[0].spines['top'].set_color('none')


axs[1].scatter(log_period, V, s=0.1, color="r")
axs[1].set(xlabel='log(Period)', ylabel='V', title='V vs log(Period)')
axs[1].grid(True)
axs[1].spines['left'].set_position('zero')
axs[1].spines['right'].set_color('none')
axs[1].spines['top'].set_color('none')
```

```python
# Curve fitting

def objective(x, m, c):
    return m*x + c

xdata = np.array(log_period)
ydata_1 = np.array(I)
ydata_2 = np.array(V)


p_opt_1, p_cov_1 = curve_fit(objective, xdata, ydata_1)
p_opt_2, p_cov_2 = curve_fit(objective, xdata, ydata_2)

fig_fit, axs = plt.subplots(1, 2, sharey=True)

axs[0].scatter(xdata, ydata_1, s=0.1, color='lightblue')
axs[0].plot(xdata, objective(xdata,*p_opt_1), color='b', label = "I = " + str(round(p_opt_1[0], 3)) + "log(P)+" + str(round(p_opt_1[1], 3)))
axs[0].legend(loc='best')
axs[0].grid(True)
axs[0].set(xlabel='log(Period)', ylabel='I', title='I vs log(Period)')
axs[0].spines['left'].set_position('zero')
axs[0].spines['right'].set_color('none')
axs[0].spines['top'].set_color('none')

axs[1].scatter(xdata, ydata_2, s=0.1, color="pink")
axs[1].plot(xdata, objective(xdata,*p_opt_2), color='r', label = "V = " + str(round(p_opt_2[0], 3)) + "log(P)+" + str(round(p_opt_2[1], 3)))
axs[1].legend(loc='best')
axs[1].grid(True)
axs[1].set(xlabel='log(Period)', ylabel='V', title='V vs log(Period)')
axs[1].spines['left'].set_position('zero')
axs[1].spines['right'].set_color('none')
axs[1].spines['top'].set_color('none')
```

```python
global_intercept_I = p_opt_1[1]
global_intercept_V = p_opt_2[1]
```

```python
# Calculating upper and lower bounds for RA and DEC using Confidence interval estimation
# i.e. 95% Cl ~ 2 SD

upper_ra = df_cleaned["RA_converted(DD)"].mean() + 2*df_cleaned["RA_converted(DD)"].std()
lower_ra = df_cleaned["RA_converted(DD)"].mean() - 2*df_cleaned["RA_converted(DD)"].std()

upper_decl = df_cleaned["Decl_converted(DD)"].mean() + 2*df_cleaned["Decl_converted(DD)"].std()
lower_decl = df_cleaned["Decl_converted(DD)"].mean() - 2*df_cleaned["Decl_converted(DD)"].std()
```

```python
# Removing the outliers

df_wo_outliers = df_cleaned[(df_cleaned["RA_converted(DD)"] < upper_ra) & (df_cleaned["RA_converted(DD)"] > lower_ra)]
df_wo_outliers = df_wo_outliers[(df_wo_outliers["Decl_converted(DD)"] < upper_decl) & (df_wo_outliers["Decl_converted(DD)"] > lower_decl)]
df_wo_outliers.reset_index(inplace=True, drop=True)

df_wo_outliers.describe()

```

```python
min_ra = df_wo_outliers["RA_converted(DD)"].min()
print(min_ra)
```

```python
# Plotting

fig_RAvsDEC, axs = plt.subplots(1, 2)
axs[1].scatter(df_wo_outliers["RA_converted(DD)"], df_wo_outliers["Decl_converted(DD)"], s = 1)
axs[1].set(xlabel="RA(DD)", ylabel="DEC(DD)", title="Without outliers")
axs[1].spines['left'].set_position('zero')
axs[1].spines['right'].set_color('none')
axs[1].spines['top'].set_color('none')
axs[1].grid(True)

axs[0].scatter(df_cleaned["RA_converted(DD)"], df_cleaned["Decl_converted(DD)"], s = 1)
axs[0].set(xlabel="RA(DD)", ylabel="DEC(DD)", title="With outliers")
axs[0].grid(True)
axs[0].spines['left'].set_position('zero')
axs[0].spines['right'].set_color('none')
axs[0].spines['top'].set_color('none')
```

```python
# Finding quantiles

quantile_ra = np.zeros(5)
quantile_decl = np.zeros(5)

for i in range(5):
    quantile_ra[i] = np.quantile(np.sort(np.array(df_wo_outliers["RA_converted(DD)"])), 0.20*(i+1))
    quantile_decl[i] = np.quantile(np.sort(np.array(df_wo_outliers["Decl_converted(DD)"])), 0.20*(i+1))

print(quantile_ra)
print(quantile_decl)

```

```python
# Plotting

fig_RAvsDEC_scatter = plt.figure(figsize=(10,10))
axs = fig_RAvsDEC_scatter.add_subplot(111)
axs.scatter(np.array(df_wo_outliers["RA_converted(DD)"]), np.array(df_wo_outliers["Decl_converted(DD)"]), s = 1)
for i in range(5):
    plt.axhline(y = quantile_decl[i], color='r', linewidth=1)
    plt.axvline(x = quantile_ra[i], color='r', linewidth=1)
    
axs.set(xlabel="RA(DD)", ylabel="DEC(DD)", title="Without outliers and with marked quantiles")

#plt.xlim([-0.5, 370])
axs.grid(True)
```

```python
bounds_ra = [df_wo_outliers["RA_converted(DD)"].min()]
for i in list(quantile_ra):
    bounds_ra.append(i)
    
bounds_decl = [df_wo_outliers["Decl_converted(DD)"].min()]
for i in list(quantile_decl):
    bounds_decl.append(i)
    

for i in bounds_ra:
    print(i)



```

```python
#Deviation

opt_i = []
opt_v = []

def objective(x, m, c):
    return m*x + c


for i in range(5):
    temp = df_wo_outliers[(df_wo_outliers["RA_converted(DD)"] < bounds_ra[i+1]) &  (df_wo_outliers["RA_converted(DD)"] > bounds_ra[i])]
    for j in range(5):
        df_patch = temp[(temp["Decl_converted(DD)"] < bounds_decl[j+1]) & (temp["Decl_converted(DD)"] > bounds_decl[j])]
        period = list(df_patch["P_1"])
        log_period = []
        for index in period:
            log_period.append(math.log(index))
            
        xdata = np.array(log_period)
        ydata1 = np.array(df_patch["I"])
        ydata2 = np.array(df_patch["V"])
        p_opt1 = curve_fit(objective, xdata, ydata1)
        p_opt2 = curve_fit(objective, xdata, ydata2)
        
        opt_i.append(p_opt1[0][1])
        opt_v.append(p_opt2[0][1])
        
#for i in range(5):
#    for j in range(5):
#        df_patch = df_wo_outliers[(df_wo_outliers["RA_converted(DD)"] < quantile_ra[i]) & (df_wo_outliers["Decl_converted(DD)"] < quantile_decl[j])]
#        
#        period = list(df_patch["P_1"])
#        log_period = []
#        for index in period:
#            log_period.append(math.log(index))
#        
#        xdata = np.array(log_period)
#        ydata1 = np.array(df_patch["I"])
#        ydata2 = np.array(df_patch["V"])
#
#        p_opt1 = curve_fit(objective, xdata, ydata1)
#        p_opt2 = curve_fit(objective, xdata, ydata2)
#        
#        opt_i.append(p_opt1[0][1])
#        opt_v.append(p_opt2[0][1])        
```

```python
D_vi = []

for i in range(25):
    temp = (opt_v[i] - global_intercept_V) - (opt_i[i] - global_intercept_I)
    D_vi.append(temp)
```
